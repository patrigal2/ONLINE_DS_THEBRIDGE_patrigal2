### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
#Ciudades mas pobladas de California
top_cities = df_cleaned.nlargest(5, 'population_total')


plt.figure(figsize=(10, 6))
colors = ['blue', 'grey', 'olive', 'brown', 'pink']
bars = plt.bar(top_cities['city'], top_cities['population_total'], color=colors, alpha=0.5)
plt.title("Ciudades ms pobladas de California")
plt.ylabel("Nmero de Habitantes (miles)")

#poner numero de habitantes encima de la barra
for bar, population in zip(bars, top_cities['population_total']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{population:,.0f}', 
             ha='center', va='bottom', fontsize=10)
```

```python
# Histograma del rea total ocupada por las ciudades
plt.figure(figsize=(10, 6))

plt.hist(df_cleaned['area_total_km2'], bins=50, color='steelblue', alpha=0.7)
plt.title('Histograma del rea Total Ocupada por las Ciudades', fontname='Bradley Hand ITC', fontsize=18)
plt.xlabel('Superficie (km)', fontname='Bradley Hand ITC', fontsize=18)
plt.ylabel('Nmero de Ciudades', fontname='Bradley Hand ITC', fontsize=18);


```

```python
# Scatter plot del rea de agua de una ciudad con su poblacin
plt.figure(figsize=(10, 6))

plt.scatter(df_cleaned['area_water_km2'], df_cleaned['population_total'], color='green', alpha=0.7)
plt.title('Poblacin por Kilmetro Cuadrado de Agua')
plt.xlabel('rea de Agua en la Ciudad (km)')
plt.ylabel('Poblacin (miles de habitantes)')


outliers = df_cleaned[df_cleaned['area_water_km2'] > 80]
for i in range(outliers.shape[0]):
    plt.annotate('Outlier', (outliers.iloc[i]['area_water_km2'], outliers.iloc[i]['population_total']),
                 xytext=(10, 10), textcoords='offset points', arrowprops=dict(arrowstyle='->', lw=1.5))


```

```python
# Boxplot de la altura de las ciudades
plt.figure(figsize=(10, 6))

plt.boxplot(df_cleaned['elevation_m'], vert=False, patch_artist=True, 
            boxprops=dict(facecolor='red', color='black'),
            medianprops=dict(color='yellow'))
plt.title('Diagrama de Caja de la Altitud de las Ciudades de California')
plt.xlabel('Altitud (m)');

```
