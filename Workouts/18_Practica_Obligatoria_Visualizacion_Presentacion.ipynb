### #1.2

Crea una función similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una función para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numérica del dataframe
3. Una columna numérica del dataframe diferente a la anterior.
4. Un columna categórica (Opcional, valor por defecto None)
5. Un valor para el tamaño de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamaño
6. Un sexto argumento "escala", que permita manipular la escala del tamaño en el caso de que el quinto argumento sea una cadena. Este argumento multiplicará a los valores de la columna indicada en el parámetro anterior (así si quiero dividir por 10000, este sexto argumento tendrá que valer 1/10000)

La función debe pintar el diagrama de dispersión de la columna numérica 1, en el eje X, frente a la columna numérica 2, en el eje Y, usando un código de colores en función de la columna categórica y con el tamaño de los puntos según el argumento correspondiente. 

Si no se diera una columna categórica como argumento, el color no debe pasarse como parámetro a la función de Seaborn o Matplotlib que uses.

EXTRA: Añade un argumento que controle si la función pinta una leyenda para colores y tamaños o un colorbar y una leyenda para los tamaños. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesión en vivo)

```python


def scatter_plot(df, col_x, col_y, col_hue=None, size_param=20, scale=1):
    if isinstance(size_param, str):
        if size_param in df.columns:
            sizes = df[size_param] * scale
        else:
            raise ValueError(f"La columna '{size_param}' no existe en el dataframe.")
    else:
        sizes = size_param

    plt.figure(figsize=(10, 6))
    sns.scatterplot(
        data=df,
        x=col_x,
        y=col_y,
        hue=col_hue,
        size=sizes,
        sizes=(sizes.min(), sizes.max() if isinstance(sizes, pd.Series) else size_param),
        palette="viridis"
    )
    plt.title(f'Diagrama de Dispersión: {col_x} vs {col_y}')
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.legend()
    plt.show()


archivo_CSV = "/Users/patri/Desktop/DATA_SCIENCE/ONLINE_DS_THEBRIDGE_patrigal2/Workouts/SPRINT 8/UNIDAD 1/california_cities.csv"
df_cleaned= pd.read_csv(archivo_CSV)

df_cleaned

```

```python
titanic = pd.read_csv("/Users/patri/Desktop/DATA_SCIENCE/ONLINE_DS_THEBRIDGE_patrigal2/Workouts/SPRINT 8/UNIDAD 1/titanic.csv")
titanic.head()
```

```python
#tasa de mortalidad

total_pasajeros = titanic.shape[0]
total_survived = titanic['survived'].sum()
total_fallecidos = total_pasajeros - total_survived
tasa_mortalidad = total_fallecidos / total_pasajeros * 100

#grafica

fig, ax = plt.subplots()
sns.barplot(x=["Supervivientes", "Fallecidos"], y=[total_survived, total_fallecidos], palette="viridis", ax=ax)
ax.set_title('Supervivientes vs Fallecidos en el Titanic')
ax.set_ylabel('Numero de Pasajeros');

```

```python
# Grafica supervivencia por sexo
fig, ax = plt.subplots()
sns.barplot(x="sex", y="survived", data=titanic, palette="viridis", ax=ax)
ax.set_title("Tasa supervivencia por sexo")
ax.set_ylabel("Tasa Supervivencia");


# Grafica supervivencia por edad
fig, ax = plt.subplots()
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', palette='viridis', ax=ax)
ax.set_title("Recuento de supervivencia por edad")
ax.set_xlabel("edad")
ax.set_ylabel("recuento");

```

```python
# Grafica supervivencia por clase
fig, ax = plt.subplots()
sns.barplot(x="pclass", y="survived", data=titanic, palette='viridis', ax=ax)
ax.set_title("Tasa supervivencia por clase")
ax.set_ylabel("tasa supervivencia");


# Grafica supervivencia por clase y sexo
fig, ax = plt.subplots()
sns.barplot(x="pclass", y="survived", hue="sex", data=titanic, palette="viridis", ax=ax)
ax.set_title("Tasa de supervivencia por clase y sexo")
ax.set_ylabel("tasa supervivencia");


# Grafica supervivencia por clase y edad
fig, ax = plt.subplots()
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', palette='viridis', ax=ax, kde=True)
ax.set_title("Recuento de supervivencia por edad y clase")
ax.set_xlabel("edad")
ax.set_ylabel("recuento");

```

```python
# Grafica supervivencia por puerto de embarque
fig, ax = plt.subplots()
sns.barplot(x="embarked", y="survived", data=titanic, palette="viridis", ax=ax)
ax.set_title("Tasa de supervivencia por puerto de embarque")
ax.set_ylabel("tasa de supervivencia");


# Graica tipo de pasaje por puerto de embarque
fig, ax = plt.subplots()
sns.countplot(x="embarked", hue='pclass', data=titanic, palette='viridis', ax=ax)
ax.set_title("Clase de los pasajeros por puerto de embarque")
ax.set_ylabel("recuento");

```
